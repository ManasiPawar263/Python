```python
#Import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the dataset
```

```python
glaxo = pd.read_csv('GLAXO.csv')
glaxo.head()
```

```python
beml = pd.read_csv('BEML.csv')
beml.head()
```

```python
#Select relevant column
```

```python
glaxo = glaxo[['Date','Close']]
beml = beml[['Date','Close']]
```

```python
glaxo
```

```python
#Data description
```

```python
glaxo.describe()
```

```python
glaxo.info()
```

```python
#Change the datatype of date column to date time
```

```python
glaxo['Date'] = pd.DatetimeIndex(glaxo['Date'])
glaxo
```

```python
glaxo.info()
```

```python
beml['Date'] = pd.DatetimeIndex(beml['Date'])
beml
```

```python
beml.info()
```

```python
glaxo = glaxo.set_index(glaxo['Date'])
glaxo
```

```python
beml = beml.set_index(beml['Date'])
beml
```

```python
glaxo.drop(columns = ['Date'], inplace = True)
glaxo
```

```python
beml.drop(columns = ['Date'], inplace = True)
beml
```

```python
#Plot closing value for both glaxo and beml
```

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = (300)
plt.plot(glaxo['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
plt.plot(beml['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
#Calculate gains
```

```python
glaxo['gain'] = glaxo['Close'].pct_change(periods = 1)
beml['gain'] = beml['Close'].pct_change(periods = 1)
```

```python
glaxo
```

```python
beml
```

```python
#Drop the null records
```

```python
glaxo.dropna(inplace = True)
beml.dropna(inplace = True)
```

```python
glaxo
```

```python
beml
```

```python
#Plot gain values
```

```python
plt.plot(glaxo['gain'])
plt.axhline(glaxo['gain'].mean(), color = 'r')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
plt.plot(beml['gain'])
plt.axhline(beml['gain'].mean(), color = 'y')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
#Subplot for destribution of glaxo and beml
```

```python
plt.subplot(1,2,1)
sns.distplot(glaxo['gain'], color = 'r', label = 'GLAXO')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.legend()
plt.title('Distribution Graph')

plt.subplot(1,2,2)
sns.distplot(beml['gain'], color = 'b', label = 'BEML')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.legend()
plt.title('Distribution Graph')
```
