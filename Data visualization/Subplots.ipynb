# SubPlots

- Subplots means group of axes that can exist in a single matplotlib figure. Subplots() function in 
  the matplotlib library, helps in creating multiple layouts of subplots. It provides control over     all the individual plots that are created.
- When analyzing the data you might want to compare multiple plots side-by-side. Matplotlib provides 
  a convinient method called subplots to do this.
- Subplots means a group of smaller axes (where each axes is a plot) that can exist together within
  a single figure. Think of a figure as canvas that holds multiple plots.

```python
a = np.linspace(1,15,40)
b = np.random.random_sample(40)
```

```python
fig,ax = plt.subplots(2,2)

ax[0,0].plot(a,b, color = 'g')
ax[0,1].scatter(a,b)
ax[1,0].boxplot(a)
ax[1,1].plot(b,a,color = 'g')
```

```python
#Read external data
```

```python
df = pd.read_csv('cars.csv')
df.head(6)
```

```python
#Generate subplot
```

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = (300)
```

```python
fig,axes = plt.subplots(2,2)

sns.distplot(df['width'], ax = axes[0,0])
sns.scatterplot(df['price'], df['engine-size'], hue = df['fuel-type'], ax = axes[0,1])
sns.kdeplot(df['highway-mpg'], ax = axes[1,0])
sns.boxplot(df['symboling'], ax = axes[1,1])

plt.tight_layout()
```
