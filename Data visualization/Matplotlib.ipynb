# Why to visualize data?

- Data visualization is the practice of translating information into a visual context, such as map or   graph, to make data easier for the human brain to understand and pull insights from. The main goal   of data visualization to make it easier to identify patterns, trends and outliers in large data
  sets.
- Better analysis
- Quick action
- Identifying patterns
- Finding errors
- Understanding the story
- Exploring business insights
- Grasping the latest trends

# What is matplotlib?

- Matplotlib is a python visualization library, majorly used for plotting 2D data.
- Matplotlib is a multi-platform data visualization library built on numpy arrays and designed to
  work with the broader SciPy stack.
- Matplotlib consists of several plots like line, bar, scatter, histogram, etc.

# Matplotlib line plot

```python
x = np.linspace(1,10,15)
y = [1,2,65,4,21,25,36,12,45,89,6,95,56,45,85]
plt.plot(x,y)
```

```python
#comparison of two dataset

x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y, 'black','--')

x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1,y1, 'grey')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y line graph')
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]
plt.plot(x,y, 'red')
plt.grid()            #for background grid 
```

- xlim: used to define limit on x variable
- ylim: used to define limit on y variable

```python
area = [1000,1520,2000,3000,5210]
price = [10,15,20,12,35]
plt.plot(area,price)
plt.grid()
plt.xlim(1000,6000)
plt.ylim(5,40)
plt.xlabel('Area of the house')
plt.ylabel('Price of the house')
plt.title('Area vs Price of the house')
```

# Matplotlib Histogram

- A histogram is a graph showing frequency distribution.
- It is a graph showing the number of observation within each given interval.

```python
import pandas as pd
```

```python
df = pd.read_csv('Automobile_data.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.hist(df['width'])
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.title('Width and Frequency table')
```

```python
width_mean = df['width'].mean()
width_median = df['width'].median()
df['width'].hist(bins = 10)
plt.axvline(width_mean, color = 'red')
plt.axvline(width_median, color = 'green')
plt.xlabel('Width')
plt.ylabel('Frequency')
```

```python
df['body-style'].hist(bins = 5, grid = False, alpha = 0.7, histtype = 'bar', ec = 'black')
```
